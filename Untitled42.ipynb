```python
"""
BreaKHis Binary Classifier - Benign vs Malignant
PyTorch | ResNet50 Transfer Learning | All Magnifications
"""

import os
import glob
import time
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
```

```python
# ─────────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────────
DATA_ROOT    = '/content/drive/MyDrive/Internship/Amdari Intern/Breast Cancer/histology_slides/breast'
IMG_SIZE     = (224, 224)
BATCH_SIZE   = 64
NUM_EPOCHS   = 10
LR_HEAD      = 1e-3   # learning rate for the new classifier head
LR_BACKBONE  = 1e-4   # learning rate for fine-tuning ResNet layers
WEIGHT_DECAY = 1e-4
DEVICE       = torch.device("cuda" if torch.cuda.is_available() else "cpu")
RANDOM_SEED  = 42

print(f"Using device: {DEVICE}")
```

```python
# ─────────────────────────────────────────────
# DATASET
# ─────────────────────────────────────────────
def load_image_paths(data_root):
    samples = []
    for label_name, label_idx in [("benign", 0), ("malignant", 1)]:
        pattern = os.path.join(data_root, label_name, "**", "*.png")
        paths = glob.glob(pattern, recursive=True)
        for p in paths:
            samples.append((p, label_idx))
    print(f"Total images found: {len(samples)}")
    print(f"  Benign:    {sum(1 for _, l in samples if l == 0)}")
    print(f"  Malignant: {sum(1 for _, l in samples if l == 1)}")
    return samples


class BreakHisDataset(Dataset):
    def __init__(self, samples, transform=None):
        self.samples   = samples
        self.transform = transform

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        image = Image.open(path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label
```

```python
# ─────────────────────────────────────────────
# TRANSFORMS
# ImageNet mean/std since ResNet50 was pretrained on ImageNet
# ─────────────────────────────────────────────
train_transforms = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std =[0.229, 0.224, 0.225]),
])

val_transforms = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std =[0.229, 0.224, 0.225]),
])
```

```python
# ─────────────────────────────────────────────
# MODEL — ResNet50 with custom head
# ─────────────────────────────────────────────
def build_model(freeze_backbone=True):
    model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)

    # Optionally freeze all backbone layers first (feature extraction phase)
    if freeze_backbone:
        for param in model.parameters():
            param.requires_grad = False

    # Replace the final fully connected layer
    in_features = model.fc.in_features
    model.fc = nn.Sequential(
        nn.Linear(in_features, 256),
        nn.ReLU(inplace=True),
        nn.Dropout(0.5),
        nn.Linear(256, 1)   # binary output
    )

    return model
```

```python
# ─────────────────────────────────────────────
# TRAINING
# ─────────────────────────────────────────────
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    total_loss, correct, total = 0.0, 0, 0
    for images, labels in loader:
        images = images.to(device)
        labels = labels.float().unsqueeze(1).to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * images.size(0)
        preds = (torch.sigmoid(outputs) > 0.5).float()
        correct += (preds == labels).sum().item()
        total   += images.size(0)
    return total_loss / total, correct / total


def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, correct, total = 0.0, 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.float().unsqueeze(1).to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item() * images.size(0)
            preds = (torch.sigmoid(outputs) > 0.5).float()
            correct += (preds == labels).sum().item()
            total   += images.size(0)
            all_preds.extend(preds.cpu().numpy().flatten())
            all_labels.extend(labels.cpu().numpy().flatten())
    return total_loss / total, correct / total, all_preds, all_labels
```

```python
# ─────────────────────────────────────────────
# PLOTTING
# ─────────────────────────────────────────────
def plot_history(train_losses, val_losses, train_accs, val_accs):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    epochs = range(1, len(train_losses) + 1)

    axes[0].plot(epochs, train_losses, label="Train Loss")
    axes[0].plot(epochs, val_losses,   label="Val Loss")
    axes[0].set_title("Loss"); axes[0].set_xlabel("Epoch")
    axes[0].legend(); axes[0].grid(True)

    axes[1].plot(epochs, train_accs, label="Train Acc")
    axes[1].plot(epochs, val_accs,   label="Val Acc")
    axes[1].set_title("Accuracy"); axes[1].set_xlabel("Epoch")
    axes[1].legend(); axes[1].grid(True)

    plt.tight_layout()
    plt.savefig("training_history.png", dpi=150)
    plt.show()
    print("Saved: training_history.png")


def plot_confusion_matrix(all_labels, all_preds):
    cm = confusion_matrix(all_labels, all_preds)
    fig, ax = plt.subplots(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Benign", "Malignant"],
                yticklabels=["Benign", "Malignant"], ax=ax)
    ax.set_xlabel("Predicted"); ax.set_ylabel("Actual")
    ax.set_title("Confusion Matrix")
    plt.tight_layout()
    plt.savefig("confusion_matrix.png", dpi=150)
    plt.show()
    print("Saved: confusion_matrix.png")
```

```python
# ─────────────────────────────────────────────
# MAIN
# ─────────────────────────────────────────────
def main():
    torch.manual_seed(RANDOM_SEED)
    np.random.seed(RANDOM_SEED)

    # Load paths
    all_samples = load_image_paths(DATA_ROOT)
    labels = [l for _, l in all_samples]

    # Stratified 70/15/15 split
    train_val, test_samples = train_test_split(
        all_samples, test_size=0.15, stratify=labels, random_state=RANDOM_SEED)
    labels_tv = [l for _, l in train_val]
    train_samples, val_samples = train_test_split(
        train_val, test_size=0.176, stratify=labels_tv, random_state=RANDOM_SEED)

    print(f"\nSplit — Train: {len(train_samples)} | Val: {len(val_samples)} | Test: {len(test_samples)}")

    # Dataloaders
    train_ds = BreakHisDataset(train_samples, transform=train_transforms)
    val_ds   = BreakHisDataset(val_samples,   transform=val_transforms)
    test_ds  = BreakHisDataset(test_samples,  transform=val_transforms)

    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

    # ── Phase 1: Train head only (backbone frozen, 5 epochs) ──
    print("\n── Phase 1: Training classifier head (backbone frozen) ──")
    model = build_model(freeze_backbone=True).to(DEVICE)
    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.Adam(model.fc.parameters(), lr=LR_HEAD, weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=5)

    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    best_val_acc = 0.0

    for epoch in range(1, 6):
        t0 = time.time()
        tr_loss, tr_acc = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE)
        vl_loss, vl_acc, _, _ = evaluate(model, val_loader, criterion, DEVICE)
        scheduler.step()
        train_losses.append(tr_loss); val_losses.append(vl_loss)
        train_accs.append(tr_acc);   val_accs.append(vl_acc)
        print(f"Epoch [{epoch:02d}/05] Train Loss: {tr_loss:.4f} Acc: {tr_acc:.4f} | "
              f"Val Loss: {vl_loss:.4f} Acc: {vl_acc:.4f} | Time: {time.time()-t0:.1f}s")
        if vl_acc > best_val_acc:
            best_val_acc = vl_acc
            torch.save(model.state_dict(), "best_model.pth")
            print(f"  ✔ Best model saved (val acc: {best_val_acc:.4f})")

    # ── Phase 2: Fine-tune full network (lower LR, 15 epochs) ──
    print("\n── Phase 2: Fine-tuning full network ──")
    for param in model.parameters():
        param.requires_grad = True

    optimizer = optim.Adam([
        {"params": model.fc.parameters(),  "lr": LR_HEAD},
        {"params": [p for name, p in model.named_parameters()
                    if not name.startswith("fc")], "lr": LR_BACKBONE},
    ], weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS - 5)

    for epoch in range(6, NUM_EPOCHS + 1):
        t0 = time.time()
        tr_loss, tr_acc = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE)
        vl_loss, vl_acc, _, _ = evaluate(model, val_loader, criterion, DEVICE)
        scheduler.step()
        train_losses.append(tr_loss); val_losses.append(vl_loss)
        train_accs.append(tr_acc);   val_accs.append(vl_acc)
        print(f"Epoch [{epoch:02d}/{NUM_EPOCHS}] Train Loss: {tr_loss:.4f} Acc: {tr_acc:.4f} | "
              f"Val Loss: {vl_loss:.4f} Acc: {vl_acc:.4f} | Time: {time.time()-t0:.1f}s")
        if vl_acc > best_val_acc:
            best_val_acc = vl_acc
            torch.save(model.state_dict(), "best_model.pth")
            print(f"  ✔ Best model saved (val acc: {best_val_acc:.4f})")

    # ── Final evaluation ──
    print("\n─── Test Set Evaluation ───")
    model.load_state_dict(torch.load("best_model.pth"))
    _, test_acc, all_preds, all_labels = evaluate(model, test_loader, criterion, DEVICE)
    print(f"Test Accuracy: {test_acc:.4f}")
    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds,
                                target_names=["Benign", "Malignant"]))

    plot_history(train_losses, val_losses, train_accs, val_accs)
    plot_confusion_matrix(all_labels, all_preds)


if __name__ == "__main__":
    main()
```

```python
"""
BreaKHis Binary Classifier - Benign vs Malignant
PyTorch | ResNet50 Transfer Learning | Class Imbalance Handling
"""

import os
import glob
import time
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms, models

# ─────────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────────
DATA_ROOT    = '/content/drive/MyDrive/Internship/Amdari Intern/Breast Cancer/histology_slides/breast'
IMG_SIZE     = (224, 224)
BATCH_SIZE   = 32
NUM_EPOCHS   = 20
LR_HEAD      = 1e-3
LR_BACKBONE  = 1e-4
WEIGHT_DECAY = 1e-4
DEVICE       = torch.device("cuda" if torch.cuda.is_available() else "cpu")
RANDOM_SEED  = 42

print(f"Using device: {DEVICE}")


# ─────────────────────────────────────────────
# DATASET
# ─────────────────────────────────────────────
def load_image_paths(data_root):
    samples = []
    for label_name, label_idx in [("benign", 0), ("malignant", 1)]:
        pattern = os.path.join(data_root, label_name, "**", "*.png")
        paths = glob.glob(pattern, recursive=True)
        for p in paths:
            samples.append((p, label_idx))
    print(f"Total images found: {len(samples)}")
    print(f"  Benign:    {sum(1 for _, l in samples if l == 0)}")
    print(f"  Malignant: {sum(1 for _, l in samples if l == 1)}")
    return samples


class BreakHisDataset(Dataset):
    def __init__(self, samples, transform=None):
        self.samples   = samples
        self.transform = transform

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        image = Image.open(path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label


# ─────────────────────────────────────────────
# TRANSFORMS
# ─────────────────────────────────────────────
train_transforms = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std =[0.229, 0.224, 0.225]),
])

val_transforms = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std =[0.229, 0.224, 0.225]),
])


# ─────────────────────────────────────────────
# CLASS IMBALANCE — two complementary strategies
# ─────────────────────────────────────────────
def compute_pos_weight(samples):
    """
    For BCEWithLogitsLoss: pos_weight = n_negative / n_positive
    Tells the loss to penalise missing a benign sample more.
    """
    n_benign    = sum(1 for _, l in samples if l == 0)
    n_malignant = sum(1 for _, l in samples if l == 1)
    pos_weight  = torch.tensor([n_benign / n_malignant], dtype=torch.float32)
    print(f"\nClass balance — Benign: {n_benign} | Malignant: {n_malignant}")
    print(f"pos_weight (loss): {pos_weight.item():.4f}")
    return pos_weight


def make_weighted_sampler(samples):
    """
    WeightedRandomSampler: over-samples the minority class (benign)
    so each batch has a ~50/50 class balance.
    """
    labels      = [l for _, l in samples]
    n_benign    = labels.count(0)
    n_malignant = labels.count(1)
    class_weights = {0: 1.0 / n_benign, 1: 1.0 / n_malignant}
    sample_weights = [class_weights[l] for l in labels]
    sampler = WeightedRandomSampler(
        weights     = sample_weights,
        num_samples = len(sample_weights),
        replacement = True
    )
    return sampler


# ─────────────────────────────────────────────
# MODEL
# ─────────────────────────────────────────────
def build_model(freeze_backbone=True):
    model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
    if freeze_backbone:
        for param in model.parameters():
            param.requires_grad = False
    in_features = model.fc.in_features
    model.fc = nn.Sequential(
        nn.Linear(in_features, 256),
        nn.ReLU(inplace=True),
        nn.Dropout(0.5),
        nn.Linear(256, 1)
    )
    return model


# ─────────────────────────────────────────────
# TRAINING
# ─────────────────────────────────────────────
def train_one_epoch(model, loader, criterion, optimizer, device):
    model.train()
    total_loss, correct, total = 0.0, 0, 0
    for images, labels in loader:
        images = images.to(device)
        labels = labels.float().unsqueeze(1).to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * images.size(0)
        preds = (torch.sigmoid(outputs) > 0.5).float()
        correct += (preds == labels).sum().item()
        total   += images.size(0)
    return total_loss / total, correct / total


def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss, correct, total = 0.0, 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for images, labels in loader:
            images = images.to(device)
            labels = labels.float().unsqueeze(1).to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item() * images.size(0)
            preds = (torch.sigmoid(outputs) > 0.5).float()
            correct += (preds == labels).sum().item()
            total   += images.size(0)
            all_preds.extend(preds.cpu().numpy().flatten())
            all_labels.extend(labels.cpu().numpy().flatten())
    return total_loss / total, correct / total, all_preds, all_labels


# ─────────────────────────────────────────────
# PLOTTING
# ─────────────────────────────────────────────
def plot_history(train_losses, val_losses, train_accs, val_accs):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    epochs = range(1, len(train_losses) + 1)
    axes[0].plot(epochs, train_losses, label="Train Loss")
    axes[0].plot(epochs, val_losses,   label="Val Loss")
    axes[0].set_title("Loss"); axes[0].set_xlabel("Epoch")
    axes[0].legend(); axes[0].grid(True)
    axes[1].plot(epochs, train_accs, label="Train Acc")
    axes[1].plot(epochs, val_accs,   label="Val Acc")
    axes[1].set_title("Accuracy"); axes[1].set_xlabel("Epoch")
    axes[1].legend(); axes[1].grid(True)
    plt.tight_layout()
    plt.savefig("training_history_balanced.png", dpi=150)
    plt.show()
    print("Saved: training_history_balanced.png")


def plot_confusion_matrix(all_labels, all_preds):
    cm = confusion_matrix(all_labels, all_preds)
    fig, ax = plt.subplots(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Benign", "Malignant"],
                yticklabels=["Benign", "Malignant"], ax=ax)
    ax.set_xlabel("Predicted"); ax.set_ylabel("Actual")
    ax.set_title("Confusion Matrix (Balanced Training)")
    plt.tight_layout()
    plt.savefig("confusion_matrix_balanced.png", dpi=150)
    plt.show()
    print("Saved: confusion_matrix_balanced.png")


# ─────────────────────────────────────────────
# MAIN
# ─────────────────────────────────────────────
def main():
    torch.manual_seed(RANDOM_SEED)
    np.random.seed(RANDOM_SEED)

    all_samples = load_image_paths(DATA_ROOT)
    labels = [l for _, l in all_samples]

    # Stratified 70/15/15 split
    train_val, test_samples = train_test_split(
        all_samples, test_size=0.15, stratify=labels, random_state=RANDOM_SEED)
    labels_tv = [l for _, l in train_val]
    train_samples, val_samples = train_test_split(
        train_val, test_size=0.176, stratify=labels_tv, random_state=RANDOM_SEED)

    print(f"\nSplit — Train: {len(train_samples)} | Val: {len(val_samples)} | Test: {len(test_samples)}")

    # ── Strategy 1: Weighted loss ──
    pos_weight = compute_pos_weight(train_samples).to(DEVICE)
    criterion  = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

    # ── Strategy 2: Weighted sampler (balanced batches) ──
    sampler    = make_weighted_sampler(train_samples)

    # Dataloaders — use sampler instead of shuffle for train
    train_ds = BreakHisDataset(train_samples, transform=train_transforms)
    val_ds   = BreakHisDataset(val_samples,   transform=val_transforms)
    test_ds  = BreakHisDataset(test_samples,  transform=val_transforms)

    train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, sampler=sampler,
                              num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False,
                              num_workers=2, pin_memory=True)
    test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False,
                              num_workers=2, pin_memory=True)

    # ── Phase 1: Train head only (5 epochs) ──
    print("\n── Phase 1: Training classifier head (backbone frozen) ──")
    model     = build_model(freeze_backbone=True).to(DEVICE)
    optimizer = optim.Adam(model.fc.parameters(), lr=LR_HEAD, weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=5)

    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    best_val_acc = 0.0

    for epoch in range(1, 6):
        t0 = time.time()
        tr_loss, tr_acc = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE)
        vl_loss, vl_acc, _, _ = evaluate(model, val_loader, criterion, DEVICE)
        scheduler.step()
        train_losses.append(tr_loss); val_losses.append(vl_loss)
        train_accs.append(tr_acc);   val_accs.append(vl_acc)
        print(f"Epoch [{epoch:02d}/05] Train Loss: {tr_loss:.4f} Acc: {tr_acc:.4f} | "
              f"Val Loss: {vl_loss:.4f} Acc: {vl_acc:.4f} | Time: {time.time()-t0:.1f}s")
        if vl_acc > best_val_acc:
            best_val_acc = vl_acc
            torch.save(model.state_dict(), "best_model_balanced.pth")
            print(f"  ✔ Best model saved (val acc: {best_val_acc:.4f})")

    # ── Phase 2: Fine-tune full network (15 epochs) ──
    print("\n── Phase 2: Fine-tuning full network ──")
    for param in model.parameters():
        param.requires_grad = True

    optimizer = optim.Adam([
        {"params": model.fc.parameters(), "lr": LR_HEAD},
        {"params": [p for name, p in model.named_parameters()
                    if not name.startswith("fc")], "lr": LR_BACKBONE},
    ], weight_decay=WEIGHT_DECAY)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS - 5)

    for epoch in range(6, NUM_EPOCHS + 1):
        t0 = time.time()
        tr_loss, tr_acc = train_one_epoch(model, train_loader, criterion, optimizer, DEVICE)
        vl_loss, vl_acc, _, _ = evaluate(model, val_loader, criterion, DEVICE)
        scheduler.step()
        train_losses.append(tr_loss); val_losses.append(vl_loss)
        train_accs.append(tr_acc);   val_accs.append(vl_acc)
        print(f"Epoch [{epoch:02d}/{NUM_EPOCHS}] Train Loss: {tr_loss:.4f} Acc: {tr_acc:.4f} | "
              f"Val Loss: {vl_loss:.4f} Acc: {vl_acc:.4f} | Time: {time.time()-t0:.1f}s")
        if vl_acc > best_val_acc:
            best_val_acc = vl_acc
            torch.save(model.state_dict(), "best_model_balanced.pth")
            print(f"  ✔ Best model saved (val acc: {best_val_acc:.4f})")

    # ── Final evaluation ──
    print("\n─── Test Set Evaluation ───")
    model.load_state_dict(torch.load("best_model_balanced.pth"))
    _, test_acc, all_preds, all_labels = evaluate(model, test_loader, criterion, DEVICE)
    print(f"Test Accuracy: {test_acc:.4f}")
    print("\nClassification Report:")
    print(classification_report(all_labels, all_preds,
                                target_names=["Benign", "Malignant"]))

    plot_history(train_losses, val_losses, train_accs, val_accs)
    plot_confusion_matrix(all_labels, all_preds)


if __name__ == "__main__":
    main()
```
